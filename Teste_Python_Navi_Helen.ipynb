```python
import math #permite o c√°lculo do fatorial
import numpy as np #permite o c√°lculo do log

iMaior = 0 #contador para se obter a posi√ß√£o do maior elemento
n=10 #posi√ß√µes do vetor
vet=[] #vetor


for i in range(0, n, 1):  #loop para as posi√ß√µes dos vetores
    f=math.factorial(i) #calculando o fatorial de i para o caso par
  
    if i%2==0: #verificando se o n√∫mero √© par
        xi=round(3**i + 7*f, 2) #c√°lculo de xi para par #round para j√° arredondar para 2 casas decimais
        print(f'{xi:.2f}') #printado resultados do loop para o caso par
    else: #caso √≠mpar
        z=np.log(i).item()  #calculando o logaritmo neperiano do n√∫mero i
        xi=round(2**i + 4*z, 2) #c√°lculo de xi para caso √≠mpar
        print(f'{xi:.2f}') #printado resultados do loop para o caso √≠mpar
        
    vet.append(xi) #printando todos os elementos calculados segundo √†s ordens impostas

for index, item in enumerate(vet): #obtendo o √≠ndice com o elemento conforme itero

  if vet[index] > vet[iMaior]: # posi√ß√£o do maior elemento 
        iMaior = index

print("A posi√ß√£o do maior √©", iMaior) #aten√ß√£o: lembrar que as posi√ß√£o no python come√ßam em 0 e n√£o 1!

print('A media √©:',round(sum(vet)/len(vet),2)) 
```

```python
n = 5 #qd de alunos
i = 0 #contador iniciando em 0
notas = {} #criando um dicion√°rio vazio

while i < n:
    aluno=input('Insira nome do aluno:')
    nota = int(input('Insira a '+str(i + 1)+'¬∞ nota:'))
    notas[aluno]=nota #guardando nomes/notas no dicion√°ro
    
    i = i + 1 #loop para qtd de alunos/notas #para quando atinge a 5¬™ nota

aluno_nota = max(notas, key=notas.get) #extraindo o nome do aluno com maior nota
maior_nota = max(notas.values()) #extraindo a nota do aluno de maior nota

#Apresentando o aluno com a maior nota e sua respectiva nota
print(aluno_nota, ", ", maior_nota)
```
