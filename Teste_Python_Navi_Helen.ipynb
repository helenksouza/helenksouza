```python
c = 0 #contador iniciando em 0, sua finalidade é dizer a qtd de números números pares, múltiplos de 49 e múltiplos de 37(simultaneamente) no final
for i in range(1, 5000000+1): #último termo do range é excluso, por isso soma-se 1 para ter 5 milhões
    if i % 2 == 0 and i%49==0 and i%37==0: #verificando as condições citadas acima
        c += 1 #loop para somar números que satisfazem às condições
print(c) #qtd de números satisfatórios
```

# *Questão 2:*

image.png

```python
import math #permite o cálculo do fatorial
import numpy as np #permite o cálculo do log

iMaior = 0 #contador para se obter a posição do maior elemento
n=10 #posições do vetor
vet=[] #vetor


for i in range(0, n, 1):  #loop para as posições dos vetores
    f=math.factorial(i) #calculando o fatorial de i para o caso par
  
    if i%2==0: #verificando se o número é par
        xi=round(3**i + 7*f, 2) #cálculo de xi para par #round para já arredondar para 2 casas decimais
        print(f'{xi:.2f}') #printado resultados do loop para o caso par
    else: #caso ímpar
        z=np.log(i).item()  #calculando o logaritmo neperiano do número i
        xi=round(2**i + 4*z, 2) #cálculo de xi para caso ímpar
        print(f'{xi:.2f}') #printado resultados do loop para o caso ímpar
        
    vet.append(xi) #printando todos os elementos calculados segundo às ordens impostas

for index, item in enumerate(vet): #obtendo o índice com o elemento conforme itero

  if vet[index] > vet[iMaior]: # posição do maior elemento 
        iMaior = index

print("A posição do maior é", iMaior) #atenção: lembrar que as posição no python começam em 0 e não 1!

print('A media é:',round(sum(vet)/len(vet),2)) 
```

```python
n = 5 #qd de alunos
i = 0 #contador iniciando em 0
notas = {} #criando um dicionário vazio

while i < n:
    aluno=input('Insira nome do aluno:')
    nota = int(input('Insira a '+str(i + 1)+'° nota:'))
    notas[aluno]=nota #guardando nomes/notas no dicionáro
    
    i = i + 1 #loop para qtd de alunos/notas #para quando atinge a 5ª nota

aluno_nota = max(notas, key=notas.get) #extraindo o nome do aluno com maior nota
maior_nota = max(notas.values()) #extraindo a nota do aluno de maior nota

#Apresentando o aluno com a maior nota e sua respectiva nota
print(aluno_nota, ", ", maior_nota)
```
